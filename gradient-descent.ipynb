```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3 * x**2

def df(x):
    return 6 * x

def gradient_descent(learning_rate=0.1, epochs=20, start_x=2):
    x = start_x
    history = [x]
    loss_history = [f(x)]
    for _ in range(epochs):
        x -= learning_rate * df(x)
        history.append(x)
        loss_history.append(f(x))
    return history, loss_history

# Параметры градиентного спуска
learning_rate = 0.1
epochs = 20
start_x = 2

# Запуск алгоритма
history, loss_history = gradient_descent(learning_rate, epochs, start_x)

# Визуализация функции
x_vals = np.linspace(-2, 2, 100)
y_vals = f(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = 3x²')
plt.plot(history, [f(x) for x in history], color='red', linestyle='dashed', marker='o', label='Градиентный спуск')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Градиентный спуск для f(x) = 3x²')
plt.show()

# График сходимости
plt.figure(figsize=(10, 5))
plt.plot(range(len(loss_history)), loss_history, marker='o', linestyle='dashed', color='blue')
plt.xlabel('Итерация')
plt.ylabel('Значение функции f(x)')
plt.title('Сходимость градиентного спуска')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Функция и её градиент
def func(X, Y):
    return 3 * X**2 + 5 * Y**2

def grad(X, Y):
    dX = 6 * X
    dY = 10 * Y
    return dX, dY

# Сетка значений
X = np.linspace(0, 5, 200)
Y = np.linspace(0, 5, 200)
X, Y = np.meshgrid(X, Y)
Z = func(X, Y)

# Градиентный спуск
x, y = 4.5, 4.5  # Начальная точка
alpha = 0.1  # Скорость обучения
points = [(x, y, func(x, y))]

for _ in range(15):
    dX, dY = grad(x, y)
    x -= alpha * dX
    y -= alpha * dY
    points.append((x, y, func(x, y)))

# Визуализация
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7, antialiased=False)
points = np.array(points)
ax.plot(points[:, 0], points[:, 1], points[:, 2], 'ro-', markersize=5, label='Градиентный спуск')

ax.view_init(45, -30)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()
plt.show()
```
